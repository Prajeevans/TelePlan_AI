```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/rsrp_dataset.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.hist(figsize=(15,10))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["distance_km"],
    df["rsrp"],
    alpha=0.2
)

plt.xlabel("Distance (km)")
plt.ylabel("RSRP (dBm)")
plt.title("Distance vs RSRP")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["frequency"],
    df["rsrp"],
    alpha=0.2
)

plt.xlabel("frequency")
plt.ylabel("RSRP (dBm)")
plt.title("Distance vs RSRP")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,7))

scatter = plt.scatter(
    df["distance_km"],
    df["frequency"],
    c=df["rsrp"],
    alpha=0.6
)

plt.xlabel("Distance (km)")
plt.ylabel("Frequency (MHz)")
plt.title("Distance vs Frequency (Color = RSRP)")

plt.colorbar(scatter, label="RSRP (dBm)")

plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,8))

plt.imshow(corr)

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")

plt.show()
```

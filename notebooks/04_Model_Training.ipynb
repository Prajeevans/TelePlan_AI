```python
import pandas as pd
import numpy as np
import joblib
```

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)


from sklearn.linear_model import LinearRegression
from sklearn.ensemble import (
    RandomForestRegressor,
    GradientBoostingRegressor
)
```

```python
df = pd.read_csv("../data/telecom_dataset.csv")

df.head()
```

```python
preprocessor = joblib.load(
    "../ml/saved_models/preprocessor.pkl"
)
```

```python
X = df.drop(
    columns=["rsrp"]
)

y = df["rsrp"]
```

```python
X_processed = preprocessor.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_processed,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
linear_model = LinearRegression()

linear_model.fit(
    X_train,
    y_train
)


linear_prediction = linear_model.predict(
    X_test
)
```

```python
print("Linear Regression")

print(
    "MAE:",
    mean_absolute_error(
        y_test,
        linear_prediction
    )
)

print(
    "RMSE:",
    mean_squared_error(
        y_test,
        linear_prediction
    )**0.5
)

print(
    "R2:",
    r2_score(
        y_test,
        linear_prediction
    )
)
```

```python
rf_model = RandomForestRegressor(

    n_estimators=200,

    random_state=42,

    n_jobs=-1
)


rf_model.fit(
    X_train,
    y_train
)


rf_prediction = rf_model.predict(
    X_test
)
```

```python
print("Random Forest")

print(
    "MAE:",
    mean_absolute_error(
        y_test,
        rf_prediction
    )
)


print(
    "RMSE:",
    mean_squared_error(
        y_test,
        rf_prediction
    )**0.5
)


print(
    "R2:",
    r2_score(
        y_test,
        rf_prediction
    )
)
```

```python
gb_model = GradientBoostingRegressor(

    n_estimators=200,

    learning_rate=0.05,

    random_state=42
)


gb_model.fit(
    X_train,
    y_train
)


gb_prediction = gb_model.predict(
    X_test
)
```

```python
print("Gradient Boosting")

print(
    "MAE:",
    mean_absolute_error(
        y_test,
        gb_prediction
    )
)


print(
    "RMSE:",
    mean_squared_error(
        y_test,
        gb_prediction
    )**0.5
)


print(
    "R2:",
    r2_score(
        y_test,
        gb_prediction
    )
)
```

```python
results = pd.DataFrame({

    "Model":[
        "Linear Regression",
        "Random Forest",
        "Gradient Boosting"
    ],

    "MAE":[

        mean_absolute_error(y_test, linear_prediction),

        mean_absolute_error(y_test, rf_prediction),

        mean_absolute_error(y_test, gb_prediction)

    ],

    "RMSE":[

        mean_squared_error(y_test, linear_prediction)**0.5,

        mean_squared_error(y_test, rf_prediction)**0.5,

        mean_squared_error(y_test, gb_prediction)**0.5

    ],

    "R2":[

        r2_score(y_test, linear_prediction),

        r2_score(y_test, rf_prediction),

        r2_score(y_test, gb_prediction)

    ]

})


results
```
